```python
#!/usr/bin/env python3

import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree
import matplotlib.pyplot  as plt 

#DEFINING IRIS CONSTRUCTOR
iris=load_iris()

#LOADING IRIS DATA
data=iris.data.tolist()
target=iris.target.tolist()

#SEPERATING TRAINING & PREDICT DATA
predict_data_3=data[149]
del data[149]
del target[149]

predict_data_2=data[99]
del data[99]
del target[99]

predict_data_1=data[49]
del data[49]
del target[49]


training_target=target
training_data=data


#print(type(predict_data_1))
#print(predict_data_2)
#print(predict_data_3)
#print(training_data)

algo=tree.DecisionTreeClassifier()

trained=algo.fit(training_data,training_target)

result1 = trained.predict([predict_data_1]).tolist()
result2 = trained.predict([predict_data_2]).tolist()
result3 = trained.predict([predict_data_3]).tolist()
print(result1)
print(result2)
print(result3)
result=result1+result2+result3

#MAX SATOSA FEATURES

maxsl_s=0
maxsw_s=0
maxpl_s=0
maxpw_s=0
for i in range(0,49):
    if maxsl_s<iris.data[i][0]: #MAX SEPAL LENGTH SATOSA
        maxsl_s = iris.data[i][0]
    if maxsw_s<iris.data[i][1]: #MAX SEPAL WIDTH SATOSA
        maxsw_s = iris.data[i][1]
    if maxpl_s<iris.data[i][2]: #MAX SEPAL LENGTH SATOSA
        maxpl_s = iris.data[i][2]
    if maxsw_s<iris.data[i][3]: #MAX SEPAL LENGTH SATOSA
        maxsw_s = iris.data[i][3]
max_satosa_fetures=[maxsl_s,maxsw_s,maxpl_s,maxpw_s]
    
#MAX VERSICOLOR FEATURES
maxsl_vc=0
maxsw_vc=0
maxpl_vc=0
maxpw_vc=0

for i in range(50,99):
    if maxsl_vc<iris.data[i][0]: #MAX SEPAL LENGTH SATOSA
        maxsl_vc = iris.data[i][0]
    if maxsw_vc<iris.data[i][1]: #MAX SEPAL WIDTH SATOSA
        maxsw_vc = iris.data[i][1]
    if maxpl_vc<iris.data[i][2]: #MAX SEPAL LENGTH SATOSA
        maxpl_vc = iris.data[i][2]
    if maxsw_vc<iris.data[i][3]: #MAX SEPAL LENGTH SATOSA
        maxsw_vc = iris.data[i][3]
max_versicolor_fetures=[maxsl_vc,maxsw_vc,maxpl_vc,maxpw_vc]

#MAX VERGINICA FEATURES
maxsl_vg=0
maxsw_vg=0
maxpl_vg=0
maxpw_vg=0

for i in range(99,149):
    if maxsl_vg<iris.data[i][0]: #MAX SEPAL LENGTH SATOSA
        maxsl_vg = iris.data[i][0]
    if maxsw_vg<iris.data[i][1]: #MAX SEPAL WIDTH SATOSA
        maxsw_vg = iris.data[i][1]
    if maxpl_vg<iris.data[i][2]: #MAX SEPAL LENGTH SATOSA
        maxpl_vg = iris.data[i][2]
    if maxsw_vg<iris.data[i][3]: #MAX SEPAL LENGTH SATOSA
        maxsw_vg = iris.data[i][3]
max_verginica_fetures=[maxsl_vg,maxsw_vg,maxpl_vg,maxpw_vg]



    
        
#PLOTTING THE RESULT(BAR-PLOT & LINE-PLOT)

plt.figure(1)
plt.suptitle("FOLWER(IRIS) PREDICTION GRAPH")

#plt.subplot(131)
plt.xlabel("TEST CASE NUMBER")
plt.ylabel("FLOWER Category")
plt.bar(result,iris.target_names)

plt.figure(2,figsize=(24,10))
plt.suptitle("FOLWER(IRIS) FEATURES GRAPH")


plt.subplot(231)
plt.xlabel("FEATURES Name")
plt.ylabel("length/width(cm)")
plt.plot(iris.feature_names,max_satosa_fetures)

plt.subplot(232)
plt.xlabel("FEATURES Name")
plt.ylabel("length/width(cm)")
plt.plot(iris.feature_names,max_versicolor_fetures)

plt.subplot(233)
plt.xlabel("FEATURES Name")
plt.ylabel("length/width(cm)")
plt.plot(iris.feature_names,max_verginica_fetures)

```
