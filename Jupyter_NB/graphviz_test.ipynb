```python
#!/usr/bin/env python3

import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree
import matplotlib.pyplot  as plt 
import graphviz

#DEFINING IRIS CONSTRUCTOR
iris=load_iris()


x=[0,50,100]
#  removing  exactly one  data from each flower 
train_data=np.delete(iris.data,x,axis=0)

#  removing the same from target 
train_target=np.delete(iris.target,x)
#  testing  data  
test_data=iris.data[x]
test_target=iris.target[x]

algo=tree.DecisionTreeClassifier()

algo.fit(train_data,train_target)

predict=algo.predict(test_data)
print(predict)

out_data=tree.export_graphviz(algo,
                    out_file=None,
                    feature_names=iris.feature_names,
                    class_names=iris.target_names,
                    filled=True,
                    rounded=True
                    )
graphviz.Source(out_data)
```
