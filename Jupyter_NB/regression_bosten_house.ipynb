```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
#from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('Housing.csv') #in form of pandas data frame
price=data['price']  #in form of pandas series
size=data['lotsize']
#print(price)
size=size.values.reshape(len(size),1)   #convert pandas series into ndarray
price=price.values.reshape(len(price),1)
```

```python
size_train,size_test,price_train,price_test=train_test_split(size,price)
plt.scatter(size_test,price_test,color='red')
plt.xlabel('size')
plt.ylabel('price')
plt.title('TEST DATA')
plt.xticks(())
plt.yticks(())
regression=linear_model.LinearRegression()
regression.fit(size_train,price_train)

plt.plot(size_test,regression.predict(size_test),color='black',linewidth=3)
plt.show()
```
